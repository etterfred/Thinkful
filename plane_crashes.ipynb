```python
# import relevant libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in plane crashes csv file to the data frame 'df'
df = pd.read_csv('plane_crashes_data.csv')
print(df)
```

```python
# filter the data for years 1980 and higher
filtered_data = df[df['year']>=1980]
```

```python
# print the mean of the filtered data
print(np.mean(filtered_data))
```

```python
# print the maximum number of fatalities in the dataset
print(np.max(df['fatalities']))
```

```python
# group data by month and sum the columns
month_data = df.groupby('month').sum()
```

```python
print(month_data)
```

```python
# print the total number of crashes and print crashes where fatalities were at least 1
print(len(df['fatalities']))
print(len(df.loc[lambda df: df['fatalities'] >= 1, :]))
```
